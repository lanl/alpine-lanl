```python
import vtk
import numpy as np
import sys
import os
import math
import time
from matplotlib import pyplot as plt
```

```python
percentage = 1.0
last_entr_val = 0.0
```

```python
## Particle position local bound based analysis
entropy_data_local = np.loadtxt('smooth_hg_local_bound_1_percent/entropy_vals.txt')
key_steps_local = np.loadtxt('smooth_hg_local_bound_1_percent/key_timesteps.txt')

## percentage based selection local bound
percentage_based_tsteps = []
percentage_based_tstep_entr = []

percentage_based_tsteps.append(entropy_data_local[0,0])
percentage_based_tstep_entr.append(entropy_data_local[0,1])
last_entr_val = entropy_data_local[0,1] 

for i in range(len(entropy_data_local)):
    
    diff = abs(entropy_data_local[i,1]-last_entr_val)
    percentage_val = (last_entr_val*percentage)/100.0
    
    if diff >= percentage_val:
        percentage_based_tsteps.append(entropy_data_local[i,0])
        percentage_based_tstep_entr.append(entropy_data_local[i,1])
        last_entr_val = entropy_data_local[i,1]
        
plt.figure(figsize=(20,5))
plt.plot(entropy_data_local[:,0],entropy_data_local[:,1])
plt.scatter(percentage_based_tsteps,percentage_based_tstep_entr,c='r',s=10)      

print ('Total selected local bound: ' + str(len(percentage_based_tsteps)))
```

```python
## Particle position global fixed bound + velmag combined based analysis
entropy_data_combined_global = np.loadtxt('smooth_hg_global_velmag_1_percent/entropy_vals.txt')
key_steps_global_combined = np.loadtxt('smooth_hg_global_velmag_1_percent/key_timesteps.txt')

## percentage based selection global position bound + velmag combined
percentage_based_tsteps = []
percentage_based_tstep_entr = []

percentage_based_tsteps.append(entropy_data_combined_global[0,0])
percentage_based_tstep_entr.append(entropy_data_combined_global[0,1])
last_entr_val = entropy_data_combined_global[0,1] 

for i in range(len(entropy_data_combined_global)):
    
    diff = abs(entropy_data_combined_global[i,1]-last_entr_val)
    percentage_val = (last_entr_val*percentage)/100.0
    
    if diff >= percentage_val:
        percentage_based_tsteps.append(entropy_data_combined_global[i,0])
        percentage_based_tstep_entr.append(entropy_data_combined_global[i,1])
        last_entr_val = entropy_data_combined_global[i,1]
        
plt.figure(figsize=(20,5))
plt.plot(entropy_data_combined_global[:,0],entropy_data_combined_global[:,1])
plt.scatter(percentage_based_tsteps,percentage_based_tstep_entr,c='r',s=15)      

print ('Total selected combined: ' + str(len(percentage_based_tsteps)))
```
