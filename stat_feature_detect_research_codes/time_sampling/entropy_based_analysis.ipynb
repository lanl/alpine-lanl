```python
import vtk
import numpy as np
import sys
import os
import math
import time
from matplotlib import pyplot as plt
```

```python
percentage = 1.0
```

```python
## Particle position global fixed bound based analysis
entropy_data_global = np.loadtxt('smooth_hg_global_bound_1_percent/entropy_vals.txt')
key_steps_global = np.loadtxt('smooth_hg_global_bound_1_percent/key_timesteps.txt')
IO_global = np.loadtxt('smooth_hg_global_bound_1_percent/IO_timming_simulation.txt')
insitu_time_global = np.loadtxt('smooth_hg_global_bound_1_percent/timming_insitu.txt')
sim_time_global = np.loadtxt('smooth_hg_global_bound_1_percent/timming_simulation.txt')

## percentage based selection global bound
percentage_based_tsteps = []
percentage_based_tstep_entr = []

percentage_based_tsteps.append(entropy_data_global[0,0])
percentage_based_tstep_entr.append(entropy_data_global[0,1])
last_entr_val = entropy_data_global[0,1] 

for i in range(len(entropy_data_global)):
    
    diff = abs(entropy_data_global[i,1]-last_entr_val)
    percentage_val = (last_entr_val*percentage)/100.0
    
    if diff >= percentage_val:
        percentage_based_tsteps.append(entropy_data_global[i,0])
        percentage_based_tstep_entr.append(entropy_data_global[i,1])
        last_entr_val = entropy_data_global[i,1]
        
plt.figure(figsize=(20,5))
plt.plot(entropy_data_global[:,0],entropy_data_global[:,1])
plt.scatter(percentage_based_tsteps,percentage_based_tstep_entr,c='r',s=10)      
plt.xlabel('Time steps',fontsize=25)
plt.ylabel('Entropy',fontsize=25)

#################
## timing computation
total_IO_time = np.sum(IO_global[:,1])
total_insitu_time = np.sum(insitu_time_global[:,1])
total_sim_time = np.sum(sim_time_global[:,1]) - total_insitu_time

print ('Total time steps selected by global bound: ' + str(len(percentage_based_tsteps)))
print ('Total sim time: ' + str(total_sim_time))
print ('Total in situ time: ' + str(total_insitu_time))
print ('Total IO time: ' + str(total_IO_time))
```
