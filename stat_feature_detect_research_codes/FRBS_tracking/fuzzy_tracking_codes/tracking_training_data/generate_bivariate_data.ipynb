```python
import numpy as np 
import matplotlib.pyplot as plt
import random
```

```python
## Generate 2D multivar Gaussian data for testing fuzzy system
numpts = 100
mean = [2, 8]
matrix = [[2.5, 1.8], [1.8,2.5]] 
dist = np.random.multivariate_normal(mean, matrix, numpts)
out_label = np.zeros([numpts])

mean1 = [8, 2]
matrix1 = [[2.5, 1.3], [1.3, 2.8]] 
dist1 = np.random.multivariate_normal(mean1, matrix1, numpts)
out_label1 = np.ones([numpts])

combined = np.concatenate((dist, dist1))
combined_label = np.concatenate((out_label, out_label1))

print (np.shape(combined_label))
print (np.shape(combined))

fig, ax = plt.subplots(1, 1, figsize=(6, 4))
ax.scatter(combined[:,0],combined[:,1],c=combined_label,cmap='RdBu',alpha=0.85)
plt.setp(ax.get_xticklabels(), visible=False)
plt.setp(ax.get_yticklabels(), visible=False)
ax.tick_params(axis='both', which='both', length=0)
plt.savefig('bivariate_scatter.png')

# np.savetxt('gaussian_data_in.csv',combined,delimiter=',')
# np.savetxt('gaussian_data_out.csv',combined_label,delimiter=',')
```
